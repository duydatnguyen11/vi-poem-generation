3. Preprocessing dataset

```python
def remove_punctuation(text):
    punctuation = string.punctuation + '...' + '…'  # Thêm dấu ba chấm và dấu chấm lửng vào danh sách dấu câu
    translator = str.maketrans('', '', punctuation)  # Tạo bảng chuyển đổi
    text_without_punctuation = text.translate(translator)  # Sử dụng bảng chuyển đổi để loại bỏ dấu câu
    return text_without_punctuation

dataset['Content'] = dataset['Content'].apply(lambda x: remove_punctuation(x))
```

```python
print(dataset['Content'][0])
```

```python
def split_content(content):
    samples = []  # Initialize an empty list to store the samples

    # Split the content into different parts of the poem based on empty lines
    poem_parts = content.split('\n\n')

    # Iterate over each part of the poem
    for poem_part in poem_parts:
        # Split each part into lines
        poem_in_lines = poem_part.split('\n')

        # Check if the number of lines in the part is less than 6
        if len(poem_in_lines) < 6:
            # If it is, append the part (represented as a list of lines) to the samples list
            samples.append(poem_in_lines)

    return samples  # Return the list of samples


# Apply the split_content function to the 'Content' column of the dataset
dataset['Content'] = dataset['Content'].apply(lambda x: split_content(x))
dataset['Content'][0]
```

```python
dataset_exploded = dataset.explode('Content')
dataset_exploded.reset_index(drop=True, inplace=True)
dataset_exploded = dataset_exploded.dropna(subset=['Content'])
dataset_exploded
```

```python
 # Import label encoder
from sklearn import preprocessing

# label_encoder object knows
# how to understand word labels.
label_encoder = preprocessing.LabelEncoder()
# Encode labels in column 'species'.
dataset_exploded['Label Encoding']= label_encoder.fit_transform(dataset_exploded['Title'])

dataset_exploded
```

```python
print(dataset_exploded['Content'][1])
```

```python
dataset_exploded['Content'] = dataset_exploded['Content'].apply(lambda x: '\n'.join(x))
dataset_exploded
```

```python
npv_poem_dataset = Dataset.from_pandas(dataset_exploded)
npv_poem_dataset
```

```python
#create train_test_split
TEST_SIZE = 0.2
npv_poem_dataset = npv_poem_dataset.train_test_split(test_size=TEST_SIZE)
npv_poem_dataset
```

4. Importing Pretrained Model

```python
import torch
from transformers import GPTNeoForCausalLM, DataCollatorForLanguageModeling, Trainer, TrainingArguments, AutoTokenizer, EarlyStoppingCallback
from torch.nn import Dropout
from torch.optim import AdamW
from transformers import get_linear_schedule_with_warmup
from transformers import GPTNeoForCausalLM, GPT2Tokenizer

model     =    GPTNeoForCausalLM.from_pretrained("NlpHUST/gpt-neo-vi-small")
tokenizer =    GPT2Tokenizer.from_pretrained("NlpHUST/gpt-neo-vi-small")

# Thêm dropout vào mô hình
model.config.attention_dropout = 0.2
model.config.hidden_dropout = 0.2
for layer in model.transformer.h:
    layer.attn.dropout = Dropout(0.2)
```

```python
#Tokenizing
tokenizer.pad_token = tokenizer.eos_token
MAX_SEQ_LEN = 200

# Tokenization function
def tokenize_function(row):
    # Tokenize the 'Content' column
    return tokenizer(
        row['Content'],
        max_length=MAX_SEQ_LEN,
        padding='max_length',
        truncation=True
    )

# Tokenize the dataset
tokenized_poem_dataset = npv_poem_dataset.map(
    tokenize_function,
    batched=True,
    num_proc=4,
    remove_columns=npv_poem_dataset['train'].column_names)
```

```python
tokenizer('Tuyển tập thơ Nguyễn Phong Việt', max_length=10, padding='max_length', truncation=True)
```

```python
# Khởi tạo optimizer
optimizer = AdamW(model.parameters(), lr=1e-5, weight_decay=0.01)

# Số bước lên lịch trình (tổng số bước đào tạo)
total_steps = len(tokenized_poem_dataset['train']) * 30  # 30 epochs (tùy chọn số epochs)

# Số bước warmup (thường là 10% tổng số bước đào tạo)
num_warmup_steps = int(0.1 * total_steps)

# Khởi tạo lịch trình với warmup
scheduler = get_linear_schedule_with_warmup(
    optimizer,
    num_warmup_steps=num_warmup_steps,
    num_training_steps=total_steps
)

training_args = TrainingArguments(
    output_dir='gpt_npv_neo_poem_generation',
    save_strategy='steps',
    learning_rate=1e-5,
    num_train_epochs=30,
    weight_decay=0.01,
    logging_dir='gpt_npv_neo_poem_generation/logs',
    logging_steps=500,
    evaluation_strategy='steps',
    eval_steps=500,
    save_steps=500,
    per_device_train_batch_size=4,
    per_device_eval_batch_size=4,
    fp16=True,
    push_to_hub=True,
    load_best_model_at_end=True,
)
```

```python
# Thêm EarlyStoppingCallback vào Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_poem_dataset['train'],
    eval_dataset=tokenized_poem_dataset['test'],
    data_collator=DataCollatorForLanguageModeling(tokenizer=tokenizer, mlm=False),
    tokenizer=tokenizer,
    optimizers=(optimizer, scheduler),  # Thêm optimizer và scheduler vào Trainer
    callbacks=[EarlyStoppingCallback(early_stopping_patience=5, early_stopping_threshold=0.02)],
)

# Tiếp tục quá trình đào tạo
trainer.train()
```

```python
from transformers import GPTNeoForCausalLM, GPT2Tokenizer

tokenizer = GPT2Tokenizer.from_pretrained("duydatnguyen/gpt_npv_neo_poem_generation")
model = GPTNeoForCausalLM.from_pretrained("duydatnguyen/gpt_npv_neo_poem_generation")

model.config.attention_dropout = 0.15
model.config.hidden_dropout = 0.15
for layer in model.transformer.h:
    layer.attn.dropout = Dropout(0.15)


tokenizer.pad_token = tokenizer.eos_token
MAX_SEQ_LEN = 200

# Tokenization function
def tokenize_function(row):
    # Tokenize the 'Content' column
    return tokenizer(
        row['Content'],
        max_length=MAX_SEQ_LEN,
        padding='max_length',
        truncation=True
    )

# Tokenize the dataset
tokenized_poem_dataset = npv_poem_dataset.map(
    tokenize_function,
    batched=True,
    num_proc=4,
    remove_columns=npv_poem_dataset['train'].column_names,
)
```

```python
optimizer = AdamW(model.parameters(), lr=1e-5, weight_decay=0.1)

# Define total training steps and warmup steps
total_steps = len(tokenized_poem_dataset['train']) * 30
num_warmup_steps = int(0.1 * total_steps)

# Initialize scheduler with warmup
scheduler = get_linear_schedule_with_warmup(
    optimizer,
    num_warmup_steps=num_warmup_steps,
    num_training_steps=total_steps
)

# Define training arguments
training_args = TrainingArguments(
    output_dir='gpt_neo_poem_generation',
    save_strategy='steps',
    learning_rate=1e-5,  # Adjust learning rate
    num_train_epochs=30,
    weight_decay=0.1,  # Adjust weight decay
    logging_dir='gpt_neo_poem_generation/logs',
    logging_steps=500,
    evaluation_strategy='steps',
    eval_steps=500,
    save_steps=500,
    per_device_train_batch_size=4,
    per_device_eval_batch_size=4,
    fp16=True,
    push_to_hub=True,
    load_best_model_at_end=True,
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_poem_dataset['train'],
    eval_dataset=tokenized_poem_dataset['test'],
    data_collator=DataCollatorForLanguageModeling(tokenizer=tokenizer, mlm=False),
    tokenizer=tokenizer,
    optimizers=(optimizer, scheduler),
)

# Tiếp tục quá trình đào tạo
trainer.train()
```
